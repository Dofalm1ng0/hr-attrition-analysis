```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import (classification_report, confusion_matrix, roc_auc_score, roc_curve)
from imblearn.over_sampling import SMOTE
import shap
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_ml = pd.read_sql("SELECT * FROM employee_full", engine)

print(f'Shape: {df_ml.shape}')
print(f'Attrition rate: {df_ml["attrition"].mean()*100:.1f}%')
print(f'\nColumn types:')
print(df_ml.dtypes)
```

### Checking class distribution

```python
attrition_counts = df_ml['attrition'].value_counts()
attrition_pct = df_ml['attrition'].value_counts(normalize=True) * 100

print('Class Distribution:')
print(f'Stayed (0): {attrition_counts[0]:,}  ({attrition_pct[0]:.1f}%)')
print(f'Left (1): {attrition_counts[1]:,}  ({attrition_pct[1]:.1f}%)')
print(f'\nImbalance ratio: {attrition_counts[0]/attrition_counts[1]:.1f}:1')
```

## Feature Engineering

```python
drop_cols = [
    'attrition',          # target variable
    'employee_id',        # unique identifier - no predictive signal
    'hourly_rate',        # redundant with monthly_income
    'daily_rate',         # redundant with monthly_income
    'monthly_rate',       # different from monthly_income, less reliable
]

# Separate features and target
X = df_ml.drop(columns=drop_cols)
y = df_ml['attrition']

print(f'Features shape:{X.shape}')
print(f'Target shape:{y.shape}')
print(f'\nFeature columns ({X.shape[1]}):')
print(X.columns.tolist())
```

##  Encoding Categorical Columns

```python
# Identifying categorical columns
cat_cols = X.select_dtypes(include='object').columns.tolist()
print(f'Categorical columns to encode: {cat_cols}')

# Label encode each one
le = LabelEncoder()
X_encoded = X.copy()

for col in cat_cols:
    X_encoded[col] = le.fit_transform(X_encoded[col])
    print(f'{col}: {X[col].unique()} → {X_encoded[col].unique()}')

print(f'\nAll dtypes after encoding:')
print(X_encoded.dtypes.value_counts())
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42, stratify=y)
# preserves 84/16 ratio in both train and test


print(f'Training set: {X_train.shape[0]:,} rows')
print(f'Test set: {X_test.shape[0]:,} rows')
print(f'\nClass distribution in training set:')
print(y_train.value_counts())
print(f'\nClass distribution in test set:')
print(y_test.value_counts())
```

## Applying SMOTE to Training Data Only

```python
smote = SMOTE(random_state=42)
X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)

print('Before SMOTE:')
print(f'Stayed: {(y_train==0).sum():,}')
print(f'Left: {(y_train==1).sum():,}')

print('\nAfter SMOTE:')
print(f'Stayed:{(y_train_sm==0).sum():,}')
print(f'Left:{(y_train_sm==1).sum():,}')

print(f'\nTraining set size after SMOTE: {len(X_train_sm):,} rows')
```

## Training All Three Models

```python
models = {'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42), 
          'Random Forest': RandomForestClassifier( n_estimators=100,random_state=42),
          'XGBoost': XGBClassifier( n_estimators=100, learning_rate=0.1, max_depth=4, random_state=42, eval_metric='logloss', verbosity=0)}

results = {}

for name, model in models.items():
    model.fit(X_train_sm, y_train_sm)
    y_pred = model.predict(X_test)
    y_pred_prob = model.predict_proba(X_test)[:, 1]
    results[name] = {'model': model, 'y_pred': y_pred, 'y_prob': y_pred_prob, 'roc_auc': roc_auc_score(y_test, y_pred_prob)}
    
    print(f'\n{"="*50}')
    print(f'{name}')
    print(f'{"="*50}')
    print(classification_report(y_test, y_pred, target_names=['Stayed','Left']))
    print(f'ROC-AUC Score: {results[name]["roc_auc"]:.3f}')
```

## Comparing Models in One Clean Table

```python
comparison = pd.DataFrame([{'Model': name, 'ROC-AUC': round(r['roc_auc'], 3), 
                            'Report': classification_report( y_test, r['y_pred'], target_names=['Stayed','Left'],output_dict=True)}
                           for name, r in results.items()])

# Build clean summary
summary_rows = []
for name, r in results.items():
    report = classification_report(
        y_test, r['y_pred'],
        target_names=['Stayed','Left'],
        output_dict=True
    )
    summary_rows.append({
        'Model'           : name,
        'ROC-AUC'         : round(r['roc_auc'], 3),
        'Precision (Left)': round(report['Left']['precision'], 3),
        'Recall (Left)'   : round(report['Left']['recall'], 3),
        'F1 (Left)'       : round(report['Left']['f1-score'], 3),
        'Overall Accuracy': round(report['accuracy'], 3),
    })

summary_df = pd.DataFrame(summary_rows)
print(summary_df.to_string(index=False))
```
